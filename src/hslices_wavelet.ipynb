```python
def omega(k,l,m):
    f=-1.195e-4 # 55°S
    N=0.02
    H=7000
    omega = ((N**2*(k**2+l**2) + f**2*(m**2+1/(4*H**2)))/(k**2+l**2+m**2+1/(4*H**2)))**(0.5)
    return omega

f=-1.195e-4 # 55°S
ki = 2*np.pi / (750*1000)
li = 2*np.pi / (600*1000)
mi = 2*np.pi / (12*1000)
omega_intrinsic = omega(ki,li,mi)

print('Omega: ', omega_intrinsic)
print('f: ', f)
print('Faktor: ', 1-f**2/omega_intrinsic**2)
```
